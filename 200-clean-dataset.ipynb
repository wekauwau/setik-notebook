```python
import pandas as pd

df = pd.read_csv("/content/drive/My Drive/Colab/setik/dataset/100-remove-features.csv")
df.shape
```

```python
print("dataset missing values\n")

columns = df.columns.values.tolist();
n_miss = df.isna().sum().values.tolist();
perc = [x / df.shape[0] * 100 for x in n_miss]

column_max_len = max(len(x) for x in columns);
n_miss_max_len = max(len(x) for x in map(str, n_miss))

for i in range(len(columns)):
    print('{}{}{}'.format(
        columns[i].ljust(column_max_len + 1),
        str(n_miss[i]).ljust(n_miss_max_len + 1),
        f"{round(perc[i], 2)}%"
    ))
```
